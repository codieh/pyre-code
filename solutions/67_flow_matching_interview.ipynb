# Flow Matching Loss — 面试版

**难度：** Easy

**面试要点：** 展开 MSE 计算，不用 `.mean()` 手动求平均

```python
import torch
```

```python
# ✅ INTERVIEW

def flow_matching_loss(model_output, x0, x1, t):
    # ---- 目标速度 ----
    target = x1 - x0

    # ---- 手写 MSE ----
    # 面试版: 不用 .mean()，手动展开
    diff = model_output - target
    squared = diff * diff  # 逐元素平方

    # 手动求平均: sum / total_elements
    # squared.numel() = B * D
    loss = squared.sum() / squared.numel()

    return loss
```

```python
torch.manual_seed(0)
B, D = 16, 8
x0 = torch.randn(B, D)
x1 = torch.randn(B, D)
t  = torch.rand(B)

perfect_output = x1 - x0
loss = flow_matching_loss(perfect_output, x0, x1, t)
print(f"Perfect prediction => loss = {loss.item():.6f}  (expected 0.0)")
```

```python
from torch_judge import check
check("flow_matching")
```

## 📝 核心思路总结

1. **手写 MSE**：`sum(squared) / numel` 等价于 `.mean()`
2. **流匹配极简**：目标就是 `x1 - x0`，一行代码搞定
